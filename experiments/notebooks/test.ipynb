```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/patches-filtered.csv', index_col=['id', 'name'])
data.head()
```

```python
categories = pd.read_csv('https://raw.githubusercontent.com/SrMouraSilva/RBM/evaluates/experiments/data/plugins_categories_simplified.csv', index_col=['id'])
categories.head()
```

```python
categories_sorted = categories.sort_values(['category', 'name'])
categories_sorted['new_id'] = range(117)

del categories_sorted['category']
del categories_sorted['name']

categories_sorted.head()
```

```python
data_copy = data.copy()

for i in range(6):
    data_copy = data_copy.join(categories_sorted, on=f'plugin{i+1}', how='inner')
    del data_copy[f'plugin{i+1}']
    data_copy = data_copy.rename(columns={'new_id': f'plugin{i+1}'})

data_copy.head()
```

```python
data_copy.to_csv('/tmp/change_index.csv')
```
