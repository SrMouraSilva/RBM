```python
import pandas as pd
from glob import glob
```

```python
files = glob('../evaluate_results/grid-search/*.csv')
files[:5]
```

```python
# Read all data
data = pd.concat([pd.read_csv(file, index_col=[0]) for file in files])

# Split evaluation_method in three columns
temp_columns = ['classifier', 'x_transform', 'params_grid_search']
temp_data = data['evaluation_method'].str.split('-', n=2, expand=True)
temp_data.columns = temp_columns

# Add x_transform column into data
data['x_transform'] = temp_data['x_transform']
data['classifier'] = temp_data['classifier']
data.head(2)
```

```python
# Mean of the 5-fold (columns with start 'mean' refears to the mean of the 2-fold)
data_mean = data.groupby(['classifier', 'x_transform', 'params', 'column']).mean()
data_mean.head(8)
```

```python
# Organized as pivot table
#column = 'mean_test_accuracy'
#column = 'mean_test_hit@5'
#column = 'mean_test_mrr'
column = 'mean_test_mdcg'
#column = 'mean_test_map'

mean_by_column_pivot_table = data_mean[column]\
    .to_frame()\
    .pivot_table(index=['classifier', 'x_transform', 'params'], columns='column', values=column)

mean_by_column_pivot_table.head(12)
#best_pivot_table = best.pivot_table(index=['classifier', 'x_transform', 'params'], columns='column', values=column)
```

```python
# Max values by mean
mean_by_column_pivot_table\
    .T.mean().to_frame() \
    .sort_values(0, ascending=False) \
    .reset_index().drop_duplicates(['classifier', 'x_transform']) \
    .set_index(['classifier', 'x_transform', 'params'])
```

```python
def filter_by_column(data, column):
    frames = []
    index_columns = ['classifier', 'x_transform', 'params_grid_search']

    for file in files:
        data = pd.read_csv(file, index_col=[0])

        x = data['evaluation_method'].str.split('-', n=2, expand=True)
        x.columns = index_columns
        
        data['x_transform'] = x['x_transform']

        data_grouped = data.groupby(['params', 'column']).mean()
        #return data_grouped

        data_grouped = data_grouped.pivot_table(index='params', columns='column', values=column)

        for i in range(6):
            frame = {}

            frame['file'] = [file]
            frame['column'] = [i]
            frame['params'] = [data_grouped[i].idxmax()]
            frame['value'] = [data_grouped[i].max()]
            frame['x_transform'] = [data['x_transform'].max()]

            frames.append(frame)

            
    return pd.concat([pd.DataFrame(data=frame) for frame in frames])

result = filter_by_column(files[0:2], 'mean_test_accuracy')
result.head(5)
```
