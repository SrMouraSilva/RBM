```python
import pandas as pd
from glob import glob
```

```python
files = glob('../other_models/data_search_parameter/*.csv')
files
```

```python
frames = []

for file in files:
    data = pd.read_csv(file, index_col=[0])
    del data['split0_train_score']
    del data['split1_train_score']
    del data['std_train_score']
    del data['std_test_score']
    
    data_grouped = data.groupby(['params', 'is_test', 'column']).mean()
    data_grouped = data_grouped.pivot_table(index='params', columns='column', values='mean_test_score')

    for i in range(6):
        frame = {}
        
        frame['file'] = [file]
        frame['column'] = [i]
        frame['params'] = [data_grouped[i].idxmax()]
        frame['value'] = [data_grouped[i].max()]

        frames.append(frame)

result = pd.concat([pd.DataFrame(data=frame) for frame in frames])
result
```
